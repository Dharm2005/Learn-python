## Basic info about matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023,2024,2025,2026])
y = np.array([10,52,45,69])

plt.plot(x,y)
plt.show()
```

```python
# marker

x = np.array([2023,2024,2025,2026])
y = np.array([10,52,45,69])

line_style= dict(marker='.',markersize=10,markerfacecolor='blue',markeredgecolor='blue',linestyle='solid',linewidth=1,color='red')

plt.plot(x,y, **line_style)
plt.show()
```

```python
# multiple lines
y2 = np.array([25,40,77,25])
y3 = np.array([50,35,60,40])

plt.plot(x,y, **line_style)
plt.plot(x,y2, **line_style)
plt.plot(x,y3, **line_style)
plt.show()
```

```python
# lables

plt.title("CLASS SIZE", fontsize='20', family='Arial', fontweight='bold', color='blue')
plt.xlabel("Year")
plt.ylabel("Students")

plt.plot(x,y, **line_style)
plt.plot(x,y2, **line_style)
plt.plot(x,y3, **line_style)
plt.show()

plt.show()
```

```python
# ticks

plt.title("CLASS SIZE", fontsize='20', family='Arial', fontweight='bold', color='blue')
plt.xlabel("Year", color='gray')
plt.ylabel("Students", color='gray')

plt.tick_params(axis='both', colors='blue')
plt.xticks(x)

plt.plot(x,y, **line_style)
plt.plot(x,y2, **line_style)
plt.plot(x,y3, **line_style)
plt.show()

plt.show()
```

```python
# grid

x = np.array([1,2,3,4,5])
y = np.array([6,7,8,9,10])

plt.grid(linewidth=1, color='gray', linestyle='dashed')

plt.plot(x,y)
plt.show()
```
