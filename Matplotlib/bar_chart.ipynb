## Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = np.array(["Grains", "Fruits", "Vegitables", "Protein", "Dairy", "Sweets"])
values = np.array([4,3,2,5,3,1])

plt.bar(categories, values)
plt.title("Consumption")
plt.xlabel("Foods")
plt.ylabel("Quentity")
plt.show()
```
