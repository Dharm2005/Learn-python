# Handling NULL Values

```python
import pandas as pd
import numpy as np

coffee = pd.read_csv("../warmup-data/coffee.csv")
```

```python
coffee['Price'] = 4.99
coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']

coffee.head()
```

```python
coffee.loc[[2,3],"Units Sold"] = np.nan
coffee.isna().sum()
```

```python
# use mean() to fill NaN
coffee.fillna(coffee['Units Sold'].mean())

# OR

# use interpolate() to fill NaN
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
coffee.head()

# OR

# just drop row that have NaN
coffee.dropna()
coffee = coffee.dropna(subset = ['Units Sold'])
```
