# Advance Functionality

```python
# .shift() .rank() .rolling() .cumsum()
```

```python
import pandas as pd

coffee = pd.read_csv("../warmup-data/coffee.csv")
coffee['Price'] = coffee['Coffee Type'].apply(lambda x: 3.55 if x == 'Espresso' else 5.99)
coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']
```

```python
# start with .shift()

coffee['yesterday_rev'] = coffee['Revenue'].shift(2)
coffee['per_change'] = coffee['Revenue'] / coffee['yesterday_rev'] * 100
coffee
```

```python
bios = pd.read_csv('../data/bios.csv')

# .rank()

bios['height_rank'] = bios['height_cm'].rank()
bios.sort_values(['height_rank'],ascending=False)[['name','height_cm']]
```

```python
# .cumsum()

coffee['cumulative_rev'] = coffee['Revenue'].cumsum()
coffee
```

```python
# .rolling()

latte = coffee[coffee['Coffee Type'] == "Latte"].copy()

latte['3day'] = latte['Revenue'].rolling(3).sum()
latte
```
