## Scatter Graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,1,2,3,4,5,6,7,8])
y = np.array([55,60,62,68,70,75,78,82,85,87])

x1 = np.array([0,1,2,3,4,4,5,6,8,8])
y1 = np.array([50,55,62,65,72,78,80,85,82,89])

plt.scatter(x,y, color='blue',s=40, alpha=0.7,label='CLASS A')
plt.scatter(x1,y1, color='red',s=40, alpha=0.7,label='CLASS B')

plt.legend()
plt.show()
```
