# Aggregating Data

```python
import pandas as pd

bios = pd.read_csv("../data/bios.csv")
coffee = pd.read_csv("../warmup-data/coffee.csv")
coffee['Price'] = coffee['Coffee Type'].apply(lambda x: 3.55 if x == 'Espresso' else 5.99)
coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']
coffee
```

```python
# value count

bios['born_country'].value_counts()
```

```python
# combine two condition to count

bios[bios['born_country'] == 'USA']['born_region'].value_counts().head()
```

```python
# Group by

coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold' : 'sum', 'Price' : 'mean'})
```

```python
# .pivot

pivot = coffee.pivot(columns = 'Coffee Type', index='Day', values='Revenue')
pivot
```

```python
pivot.loc['Monday','Latte']
```

```python
pivot.sum(axis=0)
```

```python
pivot.sum(axis=1)
```

```python
# group by the born_date
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios['born_year'] = bios['born_date'].dt.year
bios['born_month'] = bios['born_date'].dt.month



bios.groupby([bios['born_year'], bios['born_month']])['name'].count().reset_index().sort_values('name',ascending=False)
```
