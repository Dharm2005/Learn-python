## Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['freshman','sophomores','juniors','seniors']
value = np.array([200,300,250,275])

colors =['red','green','blue','yellow']

plt.pie(value, labels=categories, autopct='%1.1f%%', colors=colors, explode=[0,0,0,0.1], shadow=True,startangle=90)

plt.title("College")

plt.show()
```
