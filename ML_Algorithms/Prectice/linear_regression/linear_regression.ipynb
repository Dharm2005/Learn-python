## Basic Linear Regression Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./house_price.csv")
df['Area_sqft'] = pd.to_numeric(df['Area_sqft'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df.dtypes
```

```python
plt.scatter(df.Area_sqft,df.Price, marker='+', color='red')
plt.ylabel('Price($)', fontweight='bold')
plt.xlabel('Area(sqft)', fontweight='bold')
plt.grid()
plt.show()
```

```python
reg = linear_model.LinearRegression()

reg.fit(df[['Area_sqft']],df['Price'])
```

```python
print("Slope (β1):", reg.coef_[0])
print("Intercept (β0):", reg.intercept_)

reg.predict([[5000]])
```

```python
plt.scatter(df.Area_sqft, df.Price, marker='+', color='red')

# Create line
area_range = np.linspace(df.Area_sqft.min(), df.Area_sqft.max(), 1000)
predicted_prices = reg.predict(area_range.reshape(-1,1))

plt.plot(area_range, predicted_prices, color='blue')   # regression line

plt.xlabel('Area(sqft)', fontweight='bold')
plt.ylabel('Price($)', fontweight='bold')
plt.grid()
plt.show()
```

```python
reg.score(df[['Area_sqft']], df['Price'])

print(f"R² = {reg.score(df[['Area_sqft']], df['Price'])}")
```
