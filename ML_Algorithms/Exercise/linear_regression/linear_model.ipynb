## Predic sales based on advertising

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# 1. Load the CSV file

df = pd.read_csv('./advertising_sales.csv')
print(df.dtypes)
df.head()

# Question: Do the values look numeric?
# Ans: Yes, the datatype shows int64 , so the values look numeric
```

```python
# 2. Visualize the data

plt.scatter(df['Advertising_Budget'],df['Sales'], marker='x', color='red')
plt.xlabel('Advertising Budget', fontweight='bold')
plt.ylabel('Sales', fontweight='bold')
plt.grid()


# Question: Do you see a linear trend?
# Ans: Yes, it looks like the data creates the streight line, so it is easy to predict
#      other data using linear regression model
```

```python
# 3. Train a Linear Regression model

reg = linear_model.LinearRegression()
reg.fit(df[['Advertising_Budget']],df['Sales'])

print(f"Slope (β₁) : {reg.coef_[0]}")
print(f"Intercept (β₀) : {reg.intercept_}")

# Question: Does sales increase a lot for every 1 rupee increase in budget?
# Ans: No, the sales increase only 9 rupee for every 1 rupee increase in budget
```

```python
# 4. Plot regression line

plt.scatter(df['Advertising_Budget'],df['Sales'], marker='x', color='red')

# Create line

area_range = np.linspace(df.Advertising_Budget.min(),df.Advertising_Budget.max(),1000)
predicted_sales = reg.predict(area_range.reshape(-1,1))

plt.plot(area_range, predicted_sales, color='blue') # regression line

plt.xlabel('Advertising Budget', fontweight='bold')
plt.ylabel('Sales', fontweight='bold')
plt.grid()
```

```python
# 5. Make predictions

# Predict sales when advertising budget is:
# ₹2500
# ₹4500
# ₹8000

print(f"₹2500 : {reg.predict([[2500]])}")
print(f"₹4500 : {reg.predict([[4500]])}")
print(f"₹8000 : {reg.predict([[8000]])}")
```

```python
# 6. Evaluate model

print(f"R² Score : {reg.score(df[['Advertising_Budget']],df['Sales'])}")
```
