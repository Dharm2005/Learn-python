# Merging & Concatenating Data

```python
import pandas as pd
```

```python
nocs = pd.read_csv("../data/noc_regions.csv")
nocs.head()
```

```python
bios = pd.read_csv("../data/bios.csv")
bios.head()
```

```python
# we will merge the data , we get region from nocs and other data from bios

bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
```

```python
bios_new.head()
```

```python
# compare NOC_x with region

bios_new[bios_new['NOC_x'] != bios_new['region']][['name','NOC_x','region']]
```

```python
# concatenating dattaframes

usa = bios[bios['born_country'] == 'USA'].copy()
gbr = bios[bios['born_country'] == 'GBR'].copy()

new_df = pd.concat([usa,gbr])

new_df.tail()
```

```python
# another example of merging

result = pd.read_csv("../data/results.csv")

combined_df = pd.merge(result, bios, on='athlete_id', how='left')

combined_df.head()
```
