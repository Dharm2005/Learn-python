# Add / Remove Columns

```python
import pandas as pd

coffee = pd.read_csv("../warmup-data/coffee.csv")
```

```python
# adding new columns
coffee['price'] = 3.99
coffee.head()
```

```python
# adding columns based on condition

import numpy as np

coffee['price'] = np.where(coffee['Coffee Type'] == 'Latte', 5.99, 2.99)
coffee
```

```python
# Drop perticular row
coffee.drop(0)
# This is not actually modifying DataFrame , it return dataframe after droping
```

```python
# to drop columns

coffee.drop(columns=["price"])
# This is not actually modifying DataFrame , it return dataframe after droping
```

# Copy DataFrame

```python
coffee = pd.read_csv('../warmup-data/coffee.csv')

# now coffee have three cols name, type and units sold

coffee_new = coffee # this will make new dataframe like coffee as coffee_new

coffee_new['price'] = 4.99 # this will add new columns in coffee_new, but... 

coffee # coffee also have price even if we didn't created it, this is because of python use same memory set for both
```

```python
# How to avoid it
coffee = pd.read_csv('../warmup-data/coffee.csv')
coffee_new = coffee.copy() # we can make copy using this syntax
coffee_new['price'] = 4.99

coffee # coffee don't have price column
```

```python
coffee_new['revenue'] = coffee_new['price'] * coffee_new['Units Sold']

coffee_new
```

```python
# Rename columns

coffee_new = coffee_new.rename(columns={'price':'Price', 'revenue':'Revenue'})
coffee_new
```

```python
coffee = coffee_new.copy()
coffee
```

# Add / Remove Data (String Operation)

```python
bios = pd.read_csv('../data/bios.csv')
bios.head()
```

```python
# let get only first name columns

bios['first name'] = bios['name'].str.split(' ').str[0]
```

```python
bios.info()
```

```python
# Convert object to date_time datatype

bios['born_dateTime'] = pd.to_datetime(bios['born_date'])
bios.head()
```

```python
bios['born_year'] = bios['born_dateTime'].dt.year # gives year from date time

bios[['name','born_year']].head()
```

```python
# save this as new csv file

bios.to_csv('../data/new_bios.csv', index = False) # save as csv file
```

# Add Using Lambda function

```python
# new columns using lambda

bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x<155 else ('Average' if x<185 else 'Tall'))

bios[['name','height_category']].loc[5:15]
```

```python
bios.loc[bios['height_category'] == 'Short',['name','height_cm']].head()
```

```python
# custom functiion

def categorize(row):
  if row['height_cm'] < 175 and row['weight_kg'] < 75:
    return 'lightweight'
  else:
    return 'average'
  
bios['category'] = bios.apply(categorize, axis=1) # axis=1 means apply on row, 0 = columns

bios.head()
```
